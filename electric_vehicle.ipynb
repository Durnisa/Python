```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Electric_Vehicle_Population_Data.csv")
df.head(2)
```

```python
df.info()
```

```python
# Checking for Null Values
df.isnull().sum()
```

```python
df.columns
```

```python
sns.heatmap(df.isnull(),cbar=False)
```

```python
df ['Postal Code']= df['Postal Code'].fillna(0)
df ['Postal Code']= df['Postal Code'].astype(int)
# Filling Null Values
```

```python
# Distributing County and City
county=df['County'].value_counts()
state=df['State'].value_counts()
print(county)
print(state)
```

```python
# Counting Vehicle Types
Vehicle_type = df['Electric Vehicle Type'].value_counts()
print(Vehicle_type)
top_model = df['Model'].value_counts().head(10)
print(top_model)
```

```python
# Visualizing Vehicles by County
plt.figure(figsize=(10,70))
sns.countplot(y='County', data=df, order=df['County'].value_counts().index)
plt.title('Electric Vehicles Distribution by County')
plt.xlabel('Count')
plt.ylabel('County')
plt.show()
```

```python
# Trend of Electric Vehicles according to Years
plt.figure(figsize=(12,6))
df['Model Year'].value_counts().sort_index().plot(kind='line')
plt.title('Trend of Electric Vehicles Over Year')
plt.xlabel('Model Year')
plt.ylabel('Count')
plt.show()
```
